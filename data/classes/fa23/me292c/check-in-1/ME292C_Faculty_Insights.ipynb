```python
import pandas as pd
```

```python
data = pd.read_csv("ME292C_CHECKIN1_MACROANALYSIS_v2.csv")
data.columns
```

```python
# Goal Setting

gs = data[["TeamName", "SharedGoalTeamAvg", "SharedGoalClassAvg", "IndvGoalTeamAvg", "IndvGoalClassAvg", "SupportTeamAvg", "SupportClassAvg"]]
```

```python
gs = gs.drop_duplicates()
```

```python
gs["GoalSettingAvg"] = gs["SharedGoalTeamAvg"] + gs["IndvGoalTeamAvg"] + gs["SupportTeamAvg"]
```

```python
list(gs.sort_values("GoalSettingAvg", ascending=False)["TeamName"])
```

```python
gs["GoalSettingClassAvg"] = gs["SharedGoalClassAvg"] + gs["IndvGoalClassAvg"] + gs["SupportClassAvg"]
```

```python
gs["GoalSettingDiffInAvgs"] = (gs["SharedGoalTeamAvg"] - gs["SharedGoalClassAvg"]) + (gs["IndvGoalTeamAvg"] - gs["IndvGoalClassAvg"])+ (gs["SupportTeamAvg"] - gs["SupportClassAvg"])
```

```python
gs.sort_values("GoalSettingDiffInAvgs", ascending=False)
```

```python
# Teammate Contributions and feedback: Initiative

initiative = data[["FullName", "TeamName", "InitiativeMe", "InitiativeTeammateAvg", "InitiativeClassAvg"]]
```

```python
#reminder: lower scores indicate disagreement. thus, 

#a large positive difference = students may feel undervalued
# a large negative difference = students think they contribute mrore than their teammates
```

```python
initiative = initiative.drop([11,36,47])
```

```python
initiative["InitiativeDiff"] = pd.to_numeric(initiative["InitiativeTeammateAvg"]) - pd.to_numeric(initiative["InitiativeMe"])
```

```python
initiative.sort_values(["TeamName", "InitiativeDiff"])
```

```python
teams = list(initiative["TeamName"].unique())
```

```python
deviations = []
for team in teams:
    std = np.std(initiative[initiative["TeamName"] == team]["InitiativeDiff"])
    deviations.append(std)
```

```python
pd.DataFrame(deviations, teams).sort_values(0, ascending=True).to_csv("InitiativeDifferences.csv")
```

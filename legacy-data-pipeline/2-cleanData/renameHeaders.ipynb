```python
cols_to_map = ['Q2_1',
 'Q2_2',
 'Q2_3',
 'Q4_1',
 'Q4_2',
 'Q4_3',
 'Q4_4',
 'Q16_1',
 'Q16_2',
 'Q16_3',
 'Q16_4',
 'Q16_5',
 'Q16_6',
 'Q17_1',
 'Q17_2',
 'Q17_3',
 'Q17_4',
 'Q17_5',
 'Q17_6',
 'Q18_1',
 'Q18_2',
 'Q18_3',
 'Q18_4',
 'Q18_5',
 'Q18_6',
 'Q19_1',
 'Q19_2',
 'Q19_3',
 'Q19_4',
 'Q19_5',
 'Q19_6']
#mappings = {
#    "Strongly agree" : 3,
#    "Agree" : 2,
#    "Somewhat agree" : 1,
#    "Neither agree nor disagree" : 0,
 #   "Somewhat disagree" : -1,
#    "Disagree" : -2,
#    "Strongly disagree" : -3,
#    np.nan : np.nan
#}
mappings = {
    '1' : 3,
    '2' : 2,
    '3' : 1,
    '4' : 0,
    '5' : -1,
    '6' : -2,
    '7' : -3,
    np.nan : np.nan
}

for col in cols_to_map:
    data[col] = [mappings[response] for response in data[col]]
```

```python
data["TeammateNumber"]
```

```python
#MAKE ALL OTHER NUM COLUMNS NON STRINGS
cols = ['Q5_1', 'Q5_2', 'Q5_3', 'Q5_4', 'Q5_5', 'Q5_6']
for col in cols: 
    data[col] = [float(j) for j in data[col]]
```

```python
data["TeammateNumber"] = [int(j) for j in data["TeammateNumber"]]
```

```python
data.to_csv("NUS_PL1101E_CH01_RAW.csv")
```
